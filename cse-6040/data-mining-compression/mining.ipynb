```python
# Book Data
import ftplib

def list_ftp_files(host, directory):
    with ftplib.FTP(host) as ftp:
        ftp.login()  # login as anonymous user
        ftp.cwd(directory)  # change to the target directory
        files = ftp.nlst()  # list files and directories
        return files

# Usage
host = 'ftp.ibiblio.org'
directory = '/pub/docs/books/shakespeare/'
files = list_ftp_files(host, directory)
print(files)
for genre in files:
    try:
        print(list_ftp_files(host, directory+genre))
    except Exception as e:
        print(e)
```

```python
# Enron emails
import requests
import xml.etree.ElementTree as ET
import pandas as pd

# URL of the XML file
xml_url = 'https://raw.githubusercontent.com/enrondata/enrondata/master/edrm-v2.0.0/edrm.enron.email.data.set.v2.xml_files.xml'

# Fetch XML content
response = requests.get(xml_url)
xml_content = response.text

# Parse XML content
root = ET.fromstring(xml_content)

# Create an empty list to store data
data = []

# Extract data from XML and append to the list
for file_element in root.findall('.//file'):
    file_name = file_element.attrib.get('name', '')
    file_source = file_element.attrib.get('source', '')
    file_mtime = file_element.find('mtime').text if file_element.find('mtime') is not None else ''
    file_size = file_element.find('size').text if file_element.find('size') is not None else ''
    file_md5 = file_element.find('md5').text if file_element.find('md5') is not None else ''
    file_crc32 = file_element.find('crc32').text if file_element.find('crc32') is not None else ''
    file_sha1 = file_element.find('sha1').text if file_element.find('sha1') is not None else ''
    file_format = file_element.find('format').text if file_element.find('format') is not None else ''
    
    data.append({
        'file_name': file_name,
        'file_source': file_source,
        'file_mtime': file_mtime,
        'file_size': file_size,
        'file_md5': file_md5,
        'file_crc32': file_crc32,
        'file_sha1': file_sha1,
        'file_format': file_format
    })

# Create a DataFrame from the list
df = pd.DataFrame(data)

# # Display the DataFrame
# print(df)
```
